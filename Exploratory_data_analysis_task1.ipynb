# Data Analysis and Visualization

```python
# State wise Sales of USA
state_sales = data.groupby('State')['Sales'].sum()
state_sales
```

```python
# State wise Sales of United States of America
plt.figure(figsize=(18,12))
sns.barplot(x=state_sales.index, y=state_sales.values)
plt.ylabel("Sales")
plt.xticks(rotation=90)
plt.show()
```

```python
# Calculating the statewise profit
profit_sum = data.groupby("State")["Profit"].sum()
profit_sum
```

```python
# State wise Profit 
plt.figure(figsize=(18,12))
sns.barplot(x=profit_sum.index, y=profit_sum.values)
plt.ylabel("Profit")
plt.xticks(rotation=90)
plt.show()
```

Analysis:

Highest profit is shown by 'California' and lowest profit is shown by 'Texas'

Top 4 profit giving states: 'California','New York','Washington'and'Michigan'

Top 4 loss giving states:'Illinois','Pennsylvania','Ohio'and'Texas'.

```python
# Calculating the citywise profit
city_profit = data.groupby('City')['Profit'].sum().sort_values(ascending = False)
city_profit
```

```python
plt.figure(figsize=(18,12))
sns.barplot(x=city_profit.index[:5], y=city_profit.values[:5])
```

```python
# Comparing Sales and Profit State  wise

profit_sales = data.groupby("State")["Profit", "Sales"].sum()
profit_sales.plot.bar(figsize=(10,6))
plt.show()
```

Inference:

Texas, Pennsylvania, and Ohio have high Sales but with loss

```python
# Distrubution of Profits of Segment across diffrent regions
plt.figure(figsize = (10,6))
sns.barplot(data = data , x = "Region",y = "Profit" ,hue = "Segment")
plt.show()
```

Inference: 

Highest profits for each region are as follows:

South= Consumer Segment

West= Corporate Segment

Central= Home Office Segment

East= Home Office Segment

Lowest profits for each region are as follows:

South= Home Office Segment

West= Home Office Segment

Central= Consumer Segment

East= Corporate Segment

```python
# Segment Vs Category
fig, ax=plt.subplots(1,3,figsize=(17,6))
sns.barplot(x='Category', y='Sales',data=data,ax=ax[0])
sns.barplot(x='Category', y='Profit',data=data,ax=ax[1])
sns.barplot(x='Category', y='Discount',data=data,ax=ax[2])
```

Highest sales in technology category

Highest profit in technology category

Highest discount in furniture category

```python
# Pie Charts of Category wise sales, Discount, Profit
category_data = data.groupby(['Category'])[['Sales', 'Discount', 'Profit']].mean()
category_data
category_data.plot.pie(subplots=True, figsize=(20, 18),autopct='%0.1f%%', labels = category_data.index)
plt.show()
```

```python
fig, ax=plt.subplots(1,3,figsize=(17,6))
sns.barplot(x='Segment', y='Sales',data=data,ax=ax[0])
sns.barplot(x='Segment', y='Profit',data=data,ax=ax[1])
sns.barplot(x='Segment', y='Discount',data=data,ax=ax[2])
```

Based on ship mode

```python
fig, ax=plt.subplots(1,3,figsize=(17,6))
sns.barplot(x='Ship Mode', y='Sales',data=data,ax=ax[0])
sns.barplot(x='Ship Mode', y='Profit',data=data,ax=ax[1])
sns.barplot(x='Ship Mode', y='Discount',data=data,ax=ax[2])
```

Inference:

Same Day Ship Mode has the highest sales.

First Class Ship mode has the highest Profit and Discount

Standard Class Ship mode has lowest sales.

Standard Class Ship mode has lowest profit.

```python
# Checking the profits of 'Sub- Category'
pivot=pd.pivot_table(data,index='Sub-Category',values='Profit')
pivot.plot(kind='bar')
plt.show()
```

```python
# Drawing Heatmap to identify correlation between numerical columns:
plt.figure(figsize=(10,8))
correlation=data.corr()
sns.heatmap(correlation,annot=True,cmap="RdYlGn")
plt.show()
```

Inference:

From the heatmap, we can see that there is positive correlation of 0.48 between Sales and Profit. Thus if Sales increases, Profit increases.

From the heatmap, we can see that there is negative correlation of -0.22 between Profit and Discount. Thus if Discount increases, most of the times Profit goes down.
